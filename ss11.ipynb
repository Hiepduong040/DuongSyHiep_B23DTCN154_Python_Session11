```python
#bai 1
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 200)
y = x**2 + 2*x

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y)
ax.set_title("Biểu đồ đầu tay của Hiệp")
ax.set_xlabel("Trục x")
ax.set_ylabel("Trục y")
ax.grid(True, linestyle="--")

plt.show()

```

```python
#Bai 4
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 200)

y1 = x
y2 = x**2
y3 = x**3

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y1, label="y = x", color="blue")
ax.plot(x, y2, label="y = x²", color="red")
ax.plot(x, y3, label="y = x³", color="green")

ax.set_title("Lineplot với Nhiều Hàm Số")
ax.set_xlabel("Trục X")
ax.set_ylabel("Giá trị Y")
ax.legend()

plt.show()
```

```python
#Bai 5
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 200)
y1 = x
y2 = x**2

rand_x = np.random.randint(0, 50, 100)
rand_y = np.random.randint(0, 50, 100)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, y1)
axes[0, 0].set_title("y = x")

axes[0, 1].plot(x, y2)
axes[0, 1].set_title("y = x²")

axes[1, 0].scatter(rand_x, rand_y)
axes[1, 0].set_title("Scatterplot 1")

axes[1, 1].scatter(rand_x, rand_y)
axes[1, 1].set_title("Scatterplot 2")

plt.tight_layout()
plt.show()
```

```python
#bai 7
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
x = np.linspace(0, 2 * np.pi, 200)

fig, ax = plt.subplots()
line, = ax.plot(x, np.sin(x))
ax.set_ylim(-1.5, 1.5)
ax.set_title("Animation: y = sin(x + t)")
ax.set_xlabel("x")
ax.set_ylabel("y")

def update(frame):
    y = np.sin(x + 0.1 * frame)
    line.set_ydata(y)
    return line,
ani = FuncAnimation(
    fig,
    update,
    frames=200,
    interval=50,  
    blit=True
)

plt.show()
```

```python
#Bai 9
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
dates = pd.date_range(start="2025-01-01", periods=30, freq="D")

temps = 25 + np.random.randn(30).cumsum() * 0.5
x = np.arange(len(temps))
coeffs = np.polyfit(x, temps, deg=1)
trend = np.polyval(coeffs, x)

diff = np.diff(temps, prepend=temps[0])
spikes_idx = np.where(np.abs(diff) > 1.5)[0]
special_dates = [
    dates[5],
    dates[15],
    dates[25]
]

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(dates, temps, label="Nhiệt độ thực tế", linewidth=2)
ax.plot(dates, trend, "--", label="Trendline", linewidth=2)

ax.scatter(dates[spikes_idx], temps[spikes_idx], 
           s=80, color="red", label="Spike (biến động mạnh)")

for d in special_dates:
    ax.axvline(d, color="green", linestyle="--", alpha=0.7)
    ax.text(d, temps.mean(), f"Mốc\n{d.date()}", 
            rotation=90, verticalalignment="center")

ax.set_title("Phân tích dữ liệu thời gian (Time Series Analysis)", fontsize=14)
ax.set_xlabel("Ngày")
ax.set_ylabel("Nhiệt độ (°C)")
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```
